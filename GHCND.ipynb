```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../cw-14-datacats/598354.csv", parse_dates=["DATE"])
```

```python
station = df[df["STATION"] == "GHCND:USC00049026"]

station_2015 = station[station["DATE"].map(lambda a: a.year==2015 and a.month==6)]
```

```python
plt.figure(figsize=(10,12), dpi=300)

plt.subplot(2,2,1)
plt.plot(station[station["TMAX"]>-300]["TMAX"])
plt.title("Maximum Temperature")
plt.xlabel("year")
plt.ylabel("temperature")

plt.subplot(2,2,2)
plt.plot(station[station["TMIN"]>-300]["TMIN"])
plt.title("Mimimum Temperature")
plt.xlabel("year")
plt.ylabel("temperature")



plt.show()
```

```python
plt.figure(figsize=(6,6), dpi=300)

max, = plt.plot(station[station["TMAX"]>-300]["TMAX"], color='firebrick')
min, = plt.plot(station[station["TMIN"]>-300]["TMIN"], color='darkturquoise')


plt.title("Temperature of Trinity River Hatchery Station")
plt.xlabel("year")
plt.ylabel("temperature $^\circ$C")
plt.legend([max, (max,min)],["Max temp", "Min temp"])





plt.show()
```

```python
plt.figure(figsize=(6,6), dpi=300)

plt.plot(station_2015["PRCP"])
plt.title("Precipitation of Trinity Stone Hatchery, June 2015")
plt.xlabel("Day")
plt.ylabel("inches")


plt.show()
```
